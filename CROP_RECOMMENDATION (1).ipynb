```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is already defined and contains your data

# Function to plot average values for a selected condition
def plot_average(condition):
    # Calculate average for each crop
    averages = df.groupby('label')[condition].mean().sort_values()
    
    # Create a bar plot
    plt.figure(figsize=(10, 6))
    averages.plot(kind='bar', color='skyblue')
    plt.title(f'Average {condition.capitalize()} Requirement for Different Crops', fontsize=16)
    plt.xlabel('Crops', fontsize=14)
    plt.ylabel(f'Average {condition.capitalize()}', fontsize=14)
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

# List of conditions to plot
conditions = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']

# Generate plots for all conditions
for condition in conditions:
    plot_average(condition)
```

```python


# Function to plot pie chart for average values of a selected condition
def plot_pie_chart(condition):
    # Calculate average for each crop
    averages = df.groupby('label')[condition].mean()
    
    # Create a pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(averages, labels=averages.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.title(f'Average {condition.capitalize()} Requirement Distribution Among Crops', fontsize=16)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

# List of conditions to plot
conditions = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']

# Generate pie charts for all conditions
for condition in conditions:
    plot_pie_chart(condition)
```

```python
# Function to plot line chart for average values of a selected condition
def plot_line_chart(condition):
    # Calculate average for each crop
    averages = df.groupby('label')[condition].mean()
    
    # Create a line chart
    plt.figure(figsize=(10, 6))
    plt.plot(averages.index, averages, marker='o')
    plt.title(f'Average {condition.capitalize()} Requirement by Crop', fontsize=16)
    plt.xlabel('Crops', fontsize=12)
    plt.ylabel(f'Average {condition.capitalize()}', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid()
    plt.tight_layout()
    plt.show()

# List of conditions to plot
conditions = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']

# Generate line charts for all conditions
for condition in conditions:
    plot_line_chart(condition)
```

```python
#Find the crops which need similar climatic conditions and mineral requirements, and club them together using Clustering Analysis

from sklearn.cluster import KMeans
x=df.drop(['label'],axis=1)
x=x.values

print(x.shape)
```

```python
#The Elbow Algorithm find the otimum number of Clusters required

plt.rcParams['figure.figsize']=(10,4)

wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i,init = 'k-means++',max_iter=300, n_init=10,random_state =0)
    km.fit(x)
    wcss.append(km.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method',fontsize=20)
plt.xlabel('No of Clusters')
plt.ylabel("wcss")
plt.show()
```

```python
#Select always the las Elbow: Hence our Cluster should be 4

#Apply Kmeans and check the Cluster of each crop: These helps Farmers to adopting similar kind of crops and yield maximum productivity
km = KMeans(n_clusters=4,init='k-means++',max_iter = 300,n_init=10,random_state=0)
y_means =km.fit_predict(x)

a=df['label']
y_means=pd.DataFrame(y_means)
z=pd.concat([y_means,a],axis=1)
z=z.rename(columns={0:'cluster'})

print(" Lets check the Results After Applying the K Means Clustering Analysis \n")
print("Crops in First Cluster:", z[z['cluster']==0]['label'].unique())
print("--------------------------------------------------------------")
print(" Crops in Second cluster:", z[z['cluster']==1]['label'].unique())
print('---------------------------------------------------------------')
print("Crops in Third cluster:", z[z['cluster'] ==2]['label'].unique())
print("----------------------------------------------------------------")
print("Crops in Forth Cluster:",z[z['cluster']==3]['label'].unique())
```

```python
#Them we train our model without labels, to avoid that our accuracy declines

y=df['label'] 
x=df.drop(['label'],axis=1)
print("Shape of x:",x.shape)
print("Shape of y:",y.shape)
```

```python
#Divide the Data into two sets; train_data and test_data
# 80% Data will be alotted into trainins set & remaining 20% will be sent into testing set

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=0)

print("The Shape of x train:",x_train.shape)
print("The Shape of x test:",x_test.shape)
print("The Shape of y train:",y_train.shape)
print("The Shape of y test:",y_test.shape)
```

```python
#Time to build a predictior model; which grides us about the best productive crop as per the climatic and minerals condition, using machine learning to train our model

#We use Logistic Regression algorythm in this case because is a probalistic model; Suitable for multiple class types; We have 22 classes in our cases

from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver = 'liblinear')

model=model.fit(x_train, y_train)

y_pred = model.predict(x_test)

#Model trained
```

```python
#Model's performance

from sklearn.metrics import confusion_matrix
```

```python
plt.rcParams['figure.figsize']=(10,10)
cm = confusion_matrix(y_test, y_pred)
sb.heatmap(cm, annot =True, cmap ='Wistia')
plt.title("Confusion Matrix for Logistic Regression",fontsize = 15)
plt.show()
```

```python
#To print the classification report (Accuracy)

from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```
